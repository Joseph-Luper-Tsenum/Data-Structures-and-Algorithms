Proof:
Induction step: n ≥ 2. 
Assume the theorem holds for n − 1 vertices.
Let G be a tree on n vertices.

Pick any leaf, v.
Let e = {v, w} be its unique edge.

Remove v and e to form graph H:
H is connected (the only paths in G with e went to/from v).
H has no cycles (they would be cycles in G, which has none).
So H is a tree with n − 1 vertices.
By the induction hypothesis, H has n − 2 edges.
Then G has (n − 2) + 1 = n − 1 edges.

#7.  Use combinations function of the itertools module and the built-in len function of Python to calculate the number of five-element subsets of the set 
{1,2…,20}
 that do not have more than one prime number in them. 

```python
#the number of five-element subsets of the set {1,2…,20} that do not have more than one prime number in them = the number of all five elements --
#--from composite number set + one element from prime number set multiply the number of 4 elements from composite number set.
from itertools import combinations
import sympy
A = set(range(1,21))             #define the {1,2...20}
B = set(sympy.primerange(0,21))  #define B = the primenumbers set [1,20]
D = A - B                        #define D = the composite number set [1,20]    
F = set(combinations(B,1))       #choose one prime number from B(the primenumbers set [1,20])
E = set(combinations(D,4))       #choose four composite numbers from D(the compose number set [1,20] )
G = set(combinations(D,5))       #choose four composite numbers from D(the compose number set [1,20] )
S = len(G) + len(E)*len(F)       #len(G) = the number of all five element from composite number set
                                 #len(F) = one element from prime number set multiply the number of 4 elements from composite number set.
S
```

#9.  Use product function of the itertools module and the built-in len function of Python to calculate the number of strings of length 10 in alphabet {A,B,C}that do not have a substring “ABA” in them. 

```python
from itertools import combinations, combinations_with_replacement, permutations, product
M = {'A','B','C'}   
R = list(product(M, repeat=10))
L = list()  #this is blank list
for rule in R:   #enter 10 elements into R, R is a kind of list.
  P = ''.join(rule)     #removing ',' from each element, and joining only characters.
  L.append(P)   #putting P into L and continue loop.
S = list()      #define S as a list = string 
for Q in L:     #for each string of L
  if 'ABA' not in Q:        #if 'ABA' is not at this string
    S.append(Q)     #delete such string and continue loop.
len(S)            #calculate the number S
```
